```python
import tensorflow_hub as hub
import tensorflow.compat.v1 as tf
tf.disable_eager_execution()
```

```python
elmo = hub.load("https://tfhub.dev/google/elmo/3").signatures["default"]
```

```python
# import spaCy's language model
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])

# function to lemmatize text
def lemmatization(texts):
    output = []
    for i in texts:
        s = [token.lemma_ for token in nlp(i)]
        output.append(' '.join(s))
    return output
```

```python
list_of_sentence= lemmatization(list_of_sentence)
list_of_sentence
```

```python
Text_1=[' '.join(i) for i in texts]
```

```python
Text_1
```

```python
Text_1=lemmatization(Text_1)
```

```python
list_of_sentence.extend(Text_1 + list_of_sentence1 + list_of_sentence2 + list_of_sentence3)
```

```python

def elmo_vectors(x):
  embeddings = elmo(tf.constant(x))["elmo"]

  with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    sess.run(tf.tables_initializer())
    # return average of ELMo features
    return sess.run(tf.reduce_mean(embeddings,1))
```

```python
elmo_train = [elmo_vectors(list_of_sentence)]
```

```python
# save elmo_train_new
pickle_out = open("elmo_train_03032019.pickle","wb")
pickle.dump(elmo_train, pickle_out)
pickle_out.close()
```
